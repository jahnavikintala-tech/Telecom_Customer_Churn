```python
import pandas as pd
df=pd.read_csv("/content/Telecom_Customer_Churn_Dataset.csv")
```

```python
import numpy as np
df.select_dtypes(include=np.number)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
a=df.select_dtypes(include=np.number)
a
```

```python
f=StandardScaler().fit_transform(a)
g=pd.DataFrame(f,columns=["Tenure_Months","Monthly_Charges","Total_Charges"])
```

```python
d=df.select_dtypes(include=["object"])
```

```python
from sklearn.preprocessing import LabelEncoder
for i in d:
  d[i]=LabelEncoder().fit_transform(d[i])
```

```python
d.head()
```

```python
h=d.drop("CustomerID",axis=1)
```

```python
df=pd.concat([g,h],axis=1)
```

```python
#Features Extraction
```

```python
y=df["Churn"]
```

```python
x=df.drop("Churn",axis=1)
```

```python
#Splitting
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
#model fitting
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model
```

```python
model.fit(x_train,y_train)
```

```python
#prediction
```

```python
y_pred=model.predict(x_test)
```

```python
#accuracyscore
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
